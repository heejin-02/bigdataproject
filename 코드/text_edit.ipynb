```python
festival_list = [
 '강릉커피축제'
,'대구약령시한방문화축제'
,'대구치맥페스티벌'
,'밀양아리랑축제'
,'봉화은어축제'
,'산청한방약초축제'
,'수원화성문화제'
,'순창장류축제'
,'시흥갯골축제'
,'안성바우덕이축제'
,'여주오곡나루축제'
,'연천구석기축제'
,'울산옹기축제'
,'원지다이내믹댄싱카니발'
,'음성품바축제'
,'인천펜타포트음악축제'
,'임실N치즈축제'
,'진안홍삼축제'
,'춘천마임축제'
,'한산모시축제'
,'횡성한우축제'
]

keywords = ['광고', '업체', '지원', '원고료', '작성'] 

for festival in festival_list:
    with open(festival+'.txt', 'r', encoding='UTF8') as file:
        lines = file.readlines()

    num_lines = len(lines)
    for i in range(0, num_lines, 3):
        title = lines[i+1].strip()
        content = lines[i+2].strip()

        found_keywords = []

        for keyword in keywords:
            if keyword in content:
                found_keywords.append(keyword)

        if found_keywords:
            keyword_str = ', '.join(found_keywords)
            print(f"Title: {title}\nContent: {content}\n다음 키워드가 포함되어 있습니다: {keyword_str}\n")
```
