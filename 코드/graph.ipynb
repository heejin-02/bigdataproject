```python
import pandas as pd
import matplotlib.pyplot as plt

# 글꼴 설정
plt.rcParams['font.family'] = 'NanumGothic'

# 엑셀 파일 불러오기
file_list = ['강릉커피축제_output.xlsx', '대구약령시한방문화축제_output.xlsx', '대구치맥페스티벌_output.xlsx', '밀양아리랑축제_output.xlsx',
             '봉화은어축제_output.xlsx', '산청한방약초축제_output.xlsx', '수원화성문화제_output.xlsx', '순창장류축제_output.xlsx',
             '시흥갯골축제_output.xlsx', '안성바우덕이축제_output.xlsx', '여주오곡나루축제_output.xlsx', '연천구석기축제_output.xlsx',
             '울산옹기축제_output.xlsx', '원주다이내믹댄싱카니발_output.xlsx', '음성품바축제_output.xlsx',
             '인천펜타포트음악축제_output.xlsx', '임실N치즈축제_output.xlsx', '진안홍삼축제_output.xlsx', '춘천마임축제_output.xlsx',
             '한산모시축제_output.xlsx', '횡성한우축제_output.xlsx']

for file_name in file_list:
    df = pd.read_excel(file_name)

    # 축제기간과 비축제기간을 분리하여 데이터 처리
    festival_df = df[df['그룹명'] == '축제기간']
    non_festival_df = df[df['그룹명'] == '비축제기간']

    # 방사형 그래프 생성
    categories = list(festival_df['구분명'])[:5]
    values_festival = list(festival_df['구분값'])[:5]
    values_non_festival = list(non_festival_df['구분값'])[:5]

    angles = [n / len(categories) * 2 * 3.14159265359 for n in range(len(categories))]
    angles += angles[:1]

    values_festival += values_festival[:1]
    values_non_festival += values_non_festival[:1]

    # 축제기간의 방사형 그래프
    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], categories, color='grey', size=8)
    plt.yticks(color='grey', size=8)
    plt.ylim(0, max(max(values_festival), max(values_non_festival)))

    ax.plot(angles, values_festival, linewidth=1, linestyle='solid', label='축제기간')
    ax.fill(angles, values_festival, 'b', alpha=0.1)

    # 비축제기간의 방사형 그래프
    ax.plot(angles, values_non_festival, linewidth=1, linestyle='solid', label='비축제기간')
    ax.fill(angles, values_non_festival, 'r', alpha=0.1)

    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.title(file_name.split('.')[0] + ' - 축제기간 vs 비축제기간', size=11, color='black', y=1.1)
    plt.savefig(file_name.split('.')[0] + '.png')
    plt.close()
```
