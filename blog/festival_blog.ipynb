```python
import urllib.request
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

festival_list = [
'강릉커피축제'
,'대구약령시한방문화축제'
,'대구치맥페스티벌'
,'밀양아리랑축제'
,'봉화은어축제'
,'산청한방약초축제'
,'수원화성문화제'
,'순창장류축제'
,'안성바우덕이축제'
,'여주오곡나루축제'
,'연천구석기축제'
,'울산옹기축제'
,'원지다이내믹댄싱카니발'
,'음성품바축제'
,'임실N치즈축제'
,'진안홍삼축제'
,'춘천마임축제'
,'한산모시축제'
,'횡성한우축제'
,'시흥갯골축제'
,'인천펜타포트음악축제'
]

#정보입력
client_id = "9VA8V3UAQBWsVm3p0RQl"
client_secret = "bN3SiKc6Oe"
display_num = '50'

for festival in festival_list:
    encText = urllib.parse.quote(festival)
    url = "https://openapi.naver.com/v1/search/blog?query=" + encText +"&display="+display_num# json 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()

    if(rescode==200):
        response_body = response.read()
    else:
        print("Error Code:" + rescode)

    body = response_body.decode('utf-8')

    #body를 나누기
    list1 = body.split('\n\t\t{\n\t\t\t')

    #naver블로그 글만 가져오기
    list1 = [i for i in list1 if 'naver' in i]

    #블로그 제목, 링크 뽑기
    titles = []
    links = []
    for i in list1:
        title = re.findall('"title":"(.*?)",\n\t\t\t"link"',i)
        link = re.findall('"link":"(.*?)",\n\t\t\t"description"',i)
        titles.append(title)
        links.append(link)

    titles = [r for i in titles for r in i]
    links = [r for i in links for r in i]

    # 링크를 다듬기 (필요없는 부분 제거 및 수정)
    blog_links = []
    for i in links:
        a = i.replace('\\','')
        b = a.replace('?Redirect=Log&logNo=','/')
        blog_links.append(b)

    df = pd.DataFrame({'제목':titles, '링크':blog_links})
    df.to_csv(festival+"_blog.csv")
    
    total_content = []
    for link in blog_links:
        url = link
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        result = soup.find_all('iframe', {'id' : 'mainFrame'})[0]['src']
        src_url = "https://blog.naver.com/" + result

        response = requests.get(src_url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # 블로그 제목
        title = soup.find_all('div', {'class' : 'se-module se-module-text se-title-text'})[0]
        total_content.append(title.text)
        total_content.append('\n')

        # 블로그 내용
        content_result = soup.find_all('div', {'class' : 'se-module se-module-text'})

        for content in content_result:
            content_text = content.find_all('p')
            for i in range(len(content_text)):
                total_content.append(content_text[i].text)
        total_content.append('\n')
        
    content_str = ' '.join(total_content)
    file = open(festival+'.txt' , 'w')
    file.write(content_str)
```
